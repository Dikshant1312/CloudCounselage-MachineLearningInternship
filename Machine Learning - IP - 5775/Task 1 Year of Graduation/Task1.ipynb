```python
academic_yr_dict ={
    '3rd Year': 3,
    'Final Year': 4,
    '2nd Year': 2,
    '1st Year': 1
}
df1['Academic Year'] = df1['Academic Year'].fillna(df1['What is your current academic year?'].map(academic_yr_dict))
df1.info()
```

```python
df1['Academic Year'].value_counts()
```

```python
df1['Academic Year'] = df1['Academic Year'].fillna(0)
df1['Academic Year'].value_counts()
```

```python
df1 = df1.drop('What is your current academic year?', axis=1)
df1.info()
```

```python
df1['Created'] = pd.to_datetime(df1['Created'])
df1['Month'] = df1['Created'].dt.month
df1['Year'] = df1['Created'].dt.year
df1.info()
```

```python
df1['Academic Year'] = df1['Academic Year'].astype(int)
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
academic_year_counts = df1['Academic Year'].value_counts().sort_index()
# print(academic_year_counts)
academic_year_frame = pd.DataFrame({
    'Academic Year': academic_year_counts.index,
    'Count': academic_year_counts.values
})
plt.figure(figsize=(10, 6))
sns.set(style="whitegrid")
barplot = sns.barplot(data=academic_year_frame, x='Academic Year', y='Count')

# text on barplot
for index, row in academic_year_frame.iterrows():
    barplot.text(x=index, y=row['Count'], s=row['Count'], color='black', ha="center")
plt.xlabel('Academic Year')
plt.ylabel('Number of Students')
plt.title('Number of Students by Academic Year')
plt.show()
```

```python
predicted_grad_year = []
for i in range(len(df1)):
    acad_year = df1.iloc[i]['Academic Year']
    curr_year = df1.iloc[i]['Year']
    if acad_year != 0:
        if acad_year == 1:
            predicted_grad_year.append(3 + curr_year)
        elif acad_year == 2:
            predicted_grad_year.append(2 + curr_year)
        elif acad_year == 3:
            predicted_grad_year.append(1 + curr_year)
        elif acad_year == 4:
            predicted_grad_year.append(curr_year)
    else:
        predicted_grad_year.append(0)

df1['Year of Graduation'] = predicted_grad_year
df1.info()
```

```python
df1['Year of Graduation'] = df1['Year of Graduation'].replace({0 : ' '})
df1['Year of Graduation'].info()
print(df1['Year of Graduation'])
```

```python
yr = df1['Year of Graduation'].value_counts()
print(yr)
year_frame = pd.DataFrame({
    'Year':yr.index,
    'Count': yr.values
})
plt.figure(figsize=(20, 20))
sns.set(style="whitegrid")
barplot = sns.barplot(data=year_frame, y='Count', x='Year',order=year_frame['Year'])
for index, row in year_frame.iterrows():
    barplot.text(x=index, y=row['Count'], s=str(row['Count']), color='black', ha="center")
plt.xlabel('graduation_Year')
plt.ylabel('number of student')
plt.title('Predicted year of graduation of student')

plt.show()
```
